# Regression ML

```python
# Feature set for avg_temp_c regression
x_reg = df.drop(columns=['avg_temp_c', 'max_temp_c', 'min_temp_c', 
                      'heat_index', 'temp_anomaly_c'])
y_reg = df['avg_temp_c']

# Encode categoricals
x_reg = pd.get_dummies(x_reg, columns=['season', 'state'], drop_first=True)
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_reg, y_reg, test_size = 0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#Building the models
```

```python
#Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

reg_LR = LinearRegression()

reg_LR.fit(x_train_scaled, y_train)

y_pred = reg_LR.predict(x_test_scaled)

s_LR = r2_score(y_test,y_pred)
print("r2_score = ",s_LR)

p_LR = reg_LR.predict(scaler.transform([[
    2017,   # year
    7.04,   # rainfall_index
    69.37,     # humidity_pct
    7.78,      # wind_speed_kmh
    4.33,      # sunshine_hrs
    0,      # el_nino_flag
    0,      # drought_flag
    0,      # season_Post-Monsoon
    0,      # season_Summer
    0,      # season_Winter
    0,      # state_Gujarat
    0,      # state_Kerala
    0,      # state_Madhya Pradesh
    0,      # state_Maharashtra
    1,      # state_Odisha
    0,      # state_Punjab
    0,      # state_Rajasthan
    0,      # state_Tamil Nadu
    0       # state_West Bengal
]]))
print("Predicted avg_temp_c:", p_LR)

#30.50
```

```python
#Decision Tree Regression

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

reg_DT = DecisionTreeRegressor()

reg_DT.fit(x_train_scaled, y_train)

y_pred = reg_DT.predict(x_test_scaled)

s_DT = r2_score(y_test,y_pred)
print("r2_score = ",s_DT)

p_DT = reg_DT.predict(scaler.transform([[
    2017,   # year
    7.04,   # rainfall_index
    69.37,     # humidity_pct
    7.78,      # wind_speed_kmh
    4.33,      # sunshine_hrs
    0,      # el_nino_flag
    0,      # drought_flag
    0,      # season_Post-Monsoon
    0,      # season_Summer
    0,      # season_Winter
    0,      # state_Gujarat
    0,      # state_Kerala
    0,      # state_Madhya Pradesh
    0,      # state_Maharashtra
    1,      # state_Odisha
    0,      # state_Punjab
    0,      # state_Rajasthan
    0,      # state_Tamil Nadu
    0       # state_West Bengal
]]))
print("Predicted avg_temp_c:", p_DT)

#30.50
```

```python
# Random Forest Regression

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

reg_RF = RandomForestRegressor()

reg_RF.fit(x_train_scaled, y_train)

y_pred = reg_RF.predict(x_test_scaled)

s_RF = r2_score(y_test,y_pred)
print("r2_score = ",s_RF)

p_RF = reg_RF.predict(scaler.transform([[
    2017,   # year
    7.04,   # rainfall_index
    69.37,     # humidity_pct
    7.78,      # wind_speed_kmh
    4.33,      # sunshine_hrs
    0,      # el_nino_flag
    0,      # drought_flag
    0,      # season_Post-Monsoon
    0,      # season_Summer
    0,      # season_Winter
    0,      # state_Gujarat
    0,      # state_Kerala
    0,      # state_Madhya Pradesh
    0,      # state_Maharashtra
    1,      # state_Odisha
    0,      # state_Punjab
    0,      # state_Rajasthan
    0,      # state_Tamil Nadu
    0       # state_West Bengal
]]))
print("Predicted avg_temp_c:", p_RF)

#30.50
```

```python
# Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

reg_GBR = GradientBoostingRegressor()

reg_GBR.fit(x_train_scaled,y_train)

y_pred = reg_GBR.predict(x_test_scaled)

s_GBR = r2_score(y_test,y_pred)
print("r2 Score = ",s_GBR)

p_GBR = reg_GBR.predict(scaler.transform([[
    2017,   # year
    7.04,   # rainfall_index
    69.37,     # humidity_pct
    7.78,      # wind_speed_kmh
    4.33,      # sunshine_hrs
    0,      # el_nino_flag
    0,      # drought_flag
    0,      # season_Post-Monsoon
    0,      # season_Summer
    0,      # season_Winter
    0,      # state_Gujarat
    0,      # state_Kerala
    0,      # state_Madhya Pradesh
    0,      # state_Maharashtra
    1,      # state_Odisha
    0,      # state_Punjab
    0,      # state_Rajasthan
    0,      # state_Tamil Nadu
    0       # state_West Bengal
]]))
print("Predicted avg_temp_c:", p_GBR)

#30.50
```

```python
# Feature set for drought_flag classification
x_cls = df.drop(columns=['drought_flag', 'rainfall_index'])
x_cls = pd.get_dummies(x_cls, columns=['season', 'state'], drop_first=True)
y_cls = df['drought_flag']
```

```python
from sklearn.model_selection import train_test_split
xc_train, xc_test, yc_train, yc_test = train_test_split(x_cls, y_cls, test_size=0.2, random_state=42, stratify=y_cls)
```

```python
scaler_cls = StandardScaler()
xc_train_scaled = scaler_cls.fit_transform(xc_train)
xc_test_scaled = scaler_cls.transform(xc_test)
```
